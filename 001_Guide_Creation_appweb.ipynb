# Déploiment du modèle dans un site web

Une fois notre page web finalisée, il est temps de l'heberger afin qu'elle soit utilisable pour tous.  Pour se faire :

**1.** il faut d'abord **créer un compte sur https://www.heroku.com/home**

**2.** Une fois inscrit et connecté, on se rend sur notre dossier contenant notre application et on la **renomme app.py**

**3.** Ensuite, rendez vous sur **https://code.visualstudio.com/download** pour installer VisualStudio Code

**4.** Une fois installé, on va sur file puis open folder et on va ouvrir notre dossier contenant notre fichier app.py  comme illustré ci-dessous :

```python
img('VS.png')
```

Et on valide

**5.** On va à l'adresse suivante **https://devcenter.heroku.com/articles/heroku-cli** on scrolle vers le bas et on utilise l'installer adapté (ici Windows 64x)

**6.** On install en laissant tout par défaut et on **redémarre Visual Studio**

**7.** Ensuite on ajoute 3 fichiers contenant les informations nécessaire au déploiment du site :

```python
img('3fichiers_heroku.png')
```

* Dans le fichier ProcFile : il contient le code que l'on entre dans le AnacondaPowerShell "**run . \app.py**"
* Dans le fichier setup.sh : pour le déploiment dans le site
* Dans le fichier requirements.txt : tout les packagings nécessaire pour le bon fonctionnement de la page (par exemple seaborn, etc.)

- **8.** Revenons sur le site heroku et allons sur le lien suivant  **https://dashboard.heroku.com/new-app** on donne un nom et on fait create app

- **9.**  on tombe sur le site **https://dashboard.heroku.com/apps/diabetemh/deploy/heroku-git** ensuite il faut suivre les étapes du guide à partir du "**\$ heroku login**". 
    - Chaque partie de code donné sur le site est à faire sur le VSCode. 
    - On retourne sur VS code, on clique sur le terminal et on note le "heroku login" :

```python
img('VS2.png')
```

**10.** On va sur l'adresse suivante **https://git-scm.com/download/win** pour installer le git (on laisse tout par defaut).
et on **referme VS pour que l'installation de git soit prit en compte**  on la relance, on retape heroku login, on se connecte sur le site dans lequel on est redirigé puis 

**11.** On retourne sur le site, et on suit les étapes dans l'encadré rouge. Toutes ces étapes sont à écrire dans le terminal de VS code

```python
img('site_heroku.png')
```

D'après l'encadré, on va taper :
- **git init** dans le terminal (pas besoin de faire le cd my project si on est deja dans le bon dossier contenant app.py que l'on a effectué dans l'étape **4**.

- **git add .**
- 
- **git commit -am "first commit"** # le text entre guillemet n'est pas important

**Si message warning/erreur après la commande** il faudra certainement configurer l'adresse mail et un nom   et taper dans le terminal de la même façon que présenté dans l'image ci-dessous :

```python
img("config_git.png")
```

Donc dans le terminal on ecrira  **git config --global user.email "VotreAdresseMail@exemple.com"** et **git config --global user.name "VotreNom"**

- **12.** On rééxecute le code **git commit -am "first commit"** et on a plus qu'à patienter le déploiment du site.
- **13.** **git push heroku master**
